```python
class Solution:
    def pattern1(self, n):
        for i in range(n):
            for j in range(n):
                print('*',end = '')
            print()

s = Solution()
s.pattern1(5)
```

```python
class Solution:
    def pattern7(self, n: int):
        for i in range(n):
            for j in range(n-i-1):
                print(' ',end = '') 
            for j in range(2*i+1):
                print('*',end = '')    
            for j in range(n-i-1):
                print(' ',end = '')
            print() 

s = Solution()
s.pattern7(2)
```

```python
class Solution:
    def pattern8(self, n: int):
        for i in range(n, 0, -1):  # loop from n down to 1
            spaces = " " * (n - i)       # leading spaces
            stars = "*" * (2 * i - 1)    # stars in each row
            print(spaces + stars)
            
s = Solution()
s.pattern8(5)
```

```python
class Solution:
    def pattern9(self, n: int):
        for i in range(n):
            for j in range(n-i-1):
                print(' ',end = '') 
            for j in range(2*i+1):
                print('*',end = '')    
            for j in range(n-i-1):
                print(' ',end = '')
            print() 

        for i in range(n, 0, -1):  # loop from n down to 1
            spaces = " " * (n - i)       # leading spaces
            stars = "*" * (2 * i - 1)    # stars in each row
            print(spaces + stars)
            
s = Solution()
s.pattern9(5)
```
