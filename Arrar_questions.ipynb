Find Largest Element in an array without sorting

```python
arr = [3, 2, 1, 5, 2]

max_val = arr[0]  

for num in arr:
    if num > max_val:
        max_val = num 

print(max_val)  
```

Second Largest Number


```python
arr = [1, 2, 4, 7, 7, 5]

max_val = float('-inf') 
second_max_val = float('-inf')

for num in arr:
    if num > max_val:
        second_max_val = max_val
        max_val = num
    elif num > second_max_val and num != max_val:  # Handles distinct second maximum
        second_max_val = num

print(second_max_val)  

min_val = float('inf') 
second_min_val = float('inf')

for num in arr:
    if num < min_val:
        second_min_val = min_val
        min_val = num
    elif num < second_min_val and num != min_val:  # Handles distinct second maximum
        second_min_val = num

print(second_min_val)  
```

Check if array is sorted

```python
def Checking_sorted_or_not(arr:list)->bool:
    for i in range(len(arr)-1):
        if arr[i] >arr[i+1]:
            return False
    return True
Checking_sorted_or_not([10,20,300,40,50,60,70])
```

Remove Duplicates in form of sorted array

```python
def Remove_Duplicates_In_Place(arr: list) -> list:
    # Loop backwards from the last element to the second element
    for i in range(len(arr) - 1, 0, -1):
        if arr[i] == arr[i - 1]:
            arr.pop(i)  # Safely removes the element at index i
    return arr

Remove_Duplicates_In_Place([10,10,20,20,300,40,50,60,70])


```
