```python
arr = [3, 2, 1, 5, 2]

max_val = arr[0]  

for num in arr:
    if num > max_val:
        max_val = num 

print(max_val)  
```

Second Largest Number


```python
arr = [1, 2, 4, 7, 7, 5]

max_val = float('-inf') 
second_max_val = float('-inf')

for num in arr:
    if num > max_val:
        second_max_val = max_val
        max_val = num
    elif num > second_max_val and num != max_val:  # Handles distinct second maximum
        second_max_val = num

print(second_max_val)  

min_val = float('inf') 
second_min_val = float('inf')

for num in arr:
    if num < min_val:
        second_min_val = min_val
        min_val = num
    elif num < second_min_val and num != min_val:  # Handles distinct second maximum
        second_min_val = num

print(second_min_val)  
```

Check if array is sorted

```python
def Checking_sorted_or_not(arr:list)->bool:
    for i in range(len(arr)-1):
        if arr[i] >arr[i+1]:
            return False
    return True
Checking_sorted_or_not([10,20,300,40,50,60,70])
```

Remove Duplicates in form of sorted array

```python
def Remove_Duplicates_In_Place(arr: list) -> list:
    # Loop backwards from the last element to the second element
    for i in range(len(arr) - 1, 0, -1):
        if arr[i] == arr[i - 1]:
            arr.pop(i)  # Safely removes the element at index i
    return arr

Remove_Duplicates_In_Place([10,10,20,20,300,40,50,60,70])


```

Move all zeros to end

```python
arr = [1,0,2,3,2,0,0,4,5,1]


def all_zeors_to_end(arr:list)-> list:
    # 1. create a temp and add all elements which are not zero
    temp = []
    for i in arr:
        if i != 0:
            temp.append(i)
    # 2. add rest of zeros at the end
    for i in range(len(arr)-len(temp)):
        temp.append(0)
    arr = temp
    return arr

    
print("here is the list of all zeros at the end \n",all_zeors_to_end(arr))
```

Linear Search


```python
arr = [1,10,23,2,2543,434,634,55,624,5436,345,21]

def Linear_Search(arr:list,target:int)->int:
    for _ in range(len(arr)):
        if arr[_] == target:
            return _
    else:
        return -1 
Linear_Search(arr,23)
```

Maximum Consecutive Order

```python
def Max_consecutive_numer(arr):
    maxi = 0
    count = 0
    for i in range(len(arr)):
        if arr[i]==1:
            count+=1
            maxi = max(count,maxi)
        else:
            count = 0
    return maxi
Max_consecutive_numer([0,1,1,1,0,0,1,1,1,1,0,0,1])
```

Find the number appear Once and rest twice using list


```python
def Longest_Subarray_Positive(arr:list):
    total_size = max(arr)
    hash_map = [0]*int(total_size+1)
    # print(hash_map,total_size) 
    for i in arr:
        hash_map[i] +=1
        # print(hash_map)
    for j in range(len(hash_map)):
        # print(f'this is loop number{j}')
        if hash_map[j]==1:
            return j 
Longest_Subarray_Positive([1,1,2,3,3,4,4])
```
