Find Largest Element in an array without sorting

```python
arr = [3, 2, 1, 5, 2]

max_val = arr[0]  

for num in arr:
    if num > max_val:
        max_val = num 

print(max_val)  
```

Second Largest Number


```python
arr = [1, 2, 4, 7, 7, 5]

max_val = float('-inf') 
second_max_val = float('-inf')

for num in arr:
    if num > max_val:
        second_max_val = max_val
        max_val = num
    elif num > second_max_val and num != max_val:  # Handles distinct second maximum
        second_max_val = num

print(second_max_val)  
```
