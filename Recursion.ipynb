```python
class Solution:
    def NnumbersSum(self, N: int):
        if N == 0:
            return 0
        return N + self.NnumbersSum(N - 1)

S = Solution()
print(S.NnumbersSum(10))
```

```python
class Solution:
    def factorial(self, n):
        if n ==1:
            return 1
        return n*self.factorial(n-1)
S = Solution()
print(S.factorial(5))
```
