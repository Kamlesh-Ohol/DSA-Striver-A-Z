Hashing the Characters


```python
arr = "Kamlesh"
hasing_table = [0] * 27
for i in range(len(arr)):
    char = arr[i].lower()
    index = ord(char) - ord('a')
    if 0 <= index < 26:
        hasing_table[index] += 1
print(hasing_table)
alphabets = []
for i in range(ord('a'), ord('z')+1):
    alphabets.append(chr(i))
print(alphabets)


for i in range(0, 256):
    print(f"ASCII {i}: {chr(i)}")
```

ASCII Whole table and its hashing


```python
arr = "Kamlesh"
hasing_table = [0] * 257
for i in range(len(arr)):
    char = arr[i]
    index = ord(char) 
    if 0 <= index < len(hasing_table):
        hasing_table[index] += 1
print(hasing_table)
ASCII_Codes = []
print("ASCII Code \t hashing number")
for i in range(0, len(hasing_table)):
    ASCII_Codes.append(chr(i))
    print(ASCII_Codes[i],"\t",hasing_table[i],"t",i)

```
