Selection Sort


```python
def selection_sort_verbose(arr: list) -> list:
    print("Initial array:", arr)
    n = len(arr)

    for i in range(n):
        print(f"\n--- Outer loop i={i} ---")
        for j in range(i+1, n):
            print(f"Comparing arr[{i}]={arr[i]} with arr[{j}]={arr[j]}")
            if arr[i] > arr[j]:
                print(f"  Swap needed: {arr[i]} > {arr[j]}")
                arr[i], arr[j] = arr[j], arr[i]
                print("  After swap:", arr)
            else:
                print(f"  No swap: {arr[i]} <= {arr[j]}")
    
    print("\nFinal sorted array:", arr)
    return arr


selection_sort_verbose([98, 76, 454, 323, 2134, 1, 12, 43, 34])
```

Bubble Sort


```python
def Bubble_sort(arr: list) -> list:
    for i in range(len(arr)):
        swaps = 0
        for j in range(0,len(arr)-i-1):
            if arr[j]>arr[j+1]:
                print(f"{arr[j],arr[j+1]}")
                print(f"Swapping cause {arr[j]} > {arr[j+1]}")
                arr[j],arr[j+1] = arr[j+1],arr[j]
                print("after swapping \n"+f"{arr[j],arr[j+1]}")
                print("*"*10)
                swaps = 1
        if swaps ==1:
            break
    return arr
Bubble_sort([98, 76, 454, 323, 2134, 1, 12, 43, 34])
```

Insertion Sort

```python
def Insertion_sort(arr: list) -> list:
    for i in range(len(arr)):
        key = arr[i]
        j = i-1
        while j>=0 and key<arr[j]:
            arr[j+1] = arr[j]
            j -=1
        arr[j+1] = key
    return arr
Insertion_sort([98, 76, 454, 323, 2134, 1, 12, 43, 34])
```

Merge Sort

```python
class Solution:
    
    def merge(self, arr, low, mid, high):
        temp = []
        left, right = low, mid + 1

    
        while left <= mid and right <= high:
            if arr[left] <= arr[right]:
                temp.append(arr[left])
                left += 1
            else:
                temp.append(arr[right])
                right += 1

    
        while left <= mid:
            temp.append(arr[left])
            left += 1

    
        while right <= high:
            temp.append(arr[right])
            right += 1

    
        for i in range(low, high + 1):
            arr[i] = temp[i - low]

    
    def mergeSort(self, arr, low, high):
        if low >= high:
            return
        mid = (low + high) // 2
        self.mergeSort(arr, low, mid)
        self.mergeSort(arr, mid + 1, high)
        self.merge(arr, low, mid, high)


# Driver code
arr = [5, 2, 8, 4, 1]
sol = Solution()
sol.mergeSort(arr, 0, len(arr) - 1)
print(*arr)
```

Quick Sort


```python
import random
def quick_sort(arr, low, high):
    if low < high:  # correct base condition
        print("\n", "#"*20, "Quick Sort Call", "#"*20)
        print("Subarray:", arr[low:high+1])
        
        pivot_index = partition(arr, low, high)
        print("Pivot placed at index", pivot_index, "->", arr)
        
        quick_sort(arr, low, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, high)
    return arr

def partition(arr, low, high):
    pivot = arr[low]  # pivot value
    i = low + 1
    j = high
    
    print("Pivot chosen:", pivot)
    
    while True:
        while i <= j and arr[i] <= pivot:
            print("i moves right:", arr[i])
            i += 1
        while i <= j and arr[j] > pivot:
            print("j moves left:", arr[j])
            j -= 1
        if i <= j:
            print(f"Swapping {arr[i]} and {arr[j]}")
            arr[i], arr[j] = arr[j], arr[i]
        else:
            break
    
    print(f"Swapping pivot {arr[low]} with {arr[j]}")
    arr[low], arr[j] = arr[j], arr[low]
    return j  # pivot index

# Example run
print(quick_sort([random.randint(0,1000) for i in range(0,10)], 0, 7))


```

Recursive Bubble sort

```python
def Bubble_sort(arr: list,n:int):
    for i in range(len(arr)):
        swaps = 0
        for j in range(0,len(arr)-i-1):
            if arr[j]>arr[j+1]:
                print(f"{arr[j],arr[j+1]}")
                print(f"Swapping cause {arr[j]} > {arr[j+1]}")
                arr[j],arr[j+1] = arr[j+1],arr[j]
                print("after swapping \n"+f"{arr[j],arr[j+1]}")
                print("*"*10)
                Bubble_sort(arr,n-1)
                swaps = 1
        if swaps ==1:
            break
    return arr
Bubble_sort([98, 76, 454, 323, 2134, 1, 12, 43, 34],10)
```

Insertion Recursive Sort


```python
def insertion_sort(arr, i, n):
    # Base case
    if i == n:
        return

    j = i
    # Move the element to the left while it's smaller than its predecessor
    while j > 0 and arr[j - 1] > arr[j]:
        # Swap
        arr[j - 1], arr[j] = arr[j], arr[j - 1]
        j -= 1

    # Recur for the next index
    insertion_sort(arr, i + 1, n)

# Driver code
arr = [13, 46, 24, 52, 20, 9]
n = len(arr)

print("Before Using Insertion Sort:")
print(arr)

insertion_sort(arr, 0, n)

print("After Using Insertion Sort:")
print(arr)
```
