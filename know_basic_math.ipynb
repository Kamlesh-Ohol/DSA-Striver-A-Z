```python
class Solution:
    def countDigit(self, n):
        step = 0
        while n!=0:
            n = n//10
            step +=1
        return step

s = Solution()
s.countDigit(123)
```

```python
class Solution:
    def reverseNumber(self, n):
        reversenumber = 0
        while n!= 0:
            digit = n%10
            reversenumber=reversenumber*10 +digit
            n = n//10
        return reversenumber

s = Solution()
print(s.reverseNumber(25))           
```

```python
class Solution:
    def isPalindrome(self, n):
        og_num = n
        reversenumber = 0
        while n!= 0:
            digit = n%10
            reversenumber=reversenumber*10 +digit
            n = n//10
            
        print(reversenumber,og_num)
        if og_num == reversenumber:
            return True
        else:
            return False
        
s = Solution()
s.isPalindrome(121)
```

```python
class Solution:
        def GCD(self, n1, n2):
            while n1>0 and n2>0:
                if n1 > n2:
                    n1 = n1 % n2
                else:
                    n2 = n2 %n1
            if n1 == 0:
                return n2
            return n1

s = Solution()
s.GCD(12134,1345342569)
    
```

```python
class Solution:
    def Armstrong(self, n):
        sum = 0
        while n!=0:
            digit = n%10
            print(n,digit)
            n = n//10
            sum = sum+(digit*digit*digit)
        return sum

s = Solution()
s.Armstrong(160)
```

```python
import math
class Solution:
    def PrintAllDivisor(self, n):
        divisors = list()
        for i in range(1,n+1):
            if n%i == 0:
                divisors.append(i)
            
        return divisors
    def PrintAllDivisorBetterMethod(self,n):
        divisors = []
        for i in range(1, int(math.sqrt(n)) + 1):
            if n % i == 0:
                divisors.append(i)
                if i != n // i:  # avoid duplicate when i is sqrt(n)
                    divisors.append(n // i)
        return sorted(divisors)
s = Solution()
s.PrintAllDivisorBetterMethod(36)
```
